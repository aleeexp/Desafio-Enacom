```python
#importando a blibioteca pandas
import pandas as pd
pd.set_option('float_format', '{:.2f}'.format)
```

```python
#importando os dados
df = pd.read_csv("Tabela investimento.csv", sep=';', encoding='utf-8')

df.head(13)
```

```python
#colocando dados extras para teste
df.loc[len(df)] = ({"Opção":"14","Descrição":"Investimento novo","Custo do investimento (R$)":350000,"Retorno esperado (R$)":350000,"Risco do investimento (R$)":"Alto"})
df.loc[len(df)] = ({"Opção":"15","Descrição":"Investimento com Erro","Custo do investimento (R$)":-620000,"Retorno esperado (R$)":500000,"Risco do investimento (R$)":"Alto"})
df.loc[len(df)] = ({"Opção":"16","Descrição":"Investimento com risco diferente","Custo do investimento (R$)":250000,"Retorno esperado (R$)":350000,"Risco do investimento (R$)":"Muito Alto"})
```

```python
#Aqui eu posso mudar minhas variaveis de acordo com a mudança cliente
capital_para_investimento = 2400000
num_min_investimento_medio = 2
num_min_investimento_alto = 1
num_min_investimento_baixo = 2
valor_max_investimento_medio = 1500000
valor_max_investimento_alto = 900000
valor_max_investimento_medio = 1200000
```

```python
#Criando uma nova coluna para descobrir a porcentagem de ganho
df['Porcentagem de ganho (%)'] = df['Retorno esperado (R$)'] / df['Custo do investimento (R$)'] *100
```

```python
#Criando mais uma coluna para ser o valor da porcentagem de ganho vezes o retorno esperado, criando assim a minha coluna base de melhores investimentos
df['Melhores Investimentos'] = df['Retorno esperado (R$)'] * df['Porcentagem de ganho (%)']
#Deixando a coluna em ordem de crescente em relação a coluna Melhores Investimentos
df = df.sort_values(by='Melhores Investimentos',ascending=False)
df
```

```python
#Aplicando uma limpeza no dataframe, deixando somente valores positivos e linhas com risco médio, baixo ou alto no dataframe
df = df[df['Melhores Investimentos'] >= 0]
filtro_medio = df[df['Risco do investimento (R$)'] == 'Médio']
filtro_baixo = df[df['Risco do investimento (R$)'] == 'Baixo']
filtro_alto = df[df['Risco do investimento (R$)'] == 'Alto']
df = pd.concat([filtro_baixo, filtro_medio, filtro_alto])


#Aplicando 3 filtros para pegar os melhores valores de cada
maiores_valores_medio = filtro_medio.nlargest(num_min_investimento_medio, 'Melhores Investimentos')
maiores_valores_baixo = filtro_baixo.nlargest(num_min_investimento_baixo, 'Melhores Investimentos')
maior_valor_alto = filtro_alto.nlargest(num_min_investimento_alto, 'Melhores Investimentos')

# Remover os valores obtidos do DataFrame original e juntar os valores obtidos em um dataframe
df = df.drop(maiores_valores_medio.index)
df = df.drop(maiores_valores_baixo.index)
df = df.drop(maior_valor_alto.index)
resultado = pd.concat([maiores_valores_medio, maiores_valores_baixo, maior_valor_alto])

#somando o valor total do custo de investimento que eu já tenho escolhido.
custo_total = resultado['Custo do investimento (R$)'].sum()
```

```python
#Criando uma lista do meu dataframe com os 5 já escolhidos.
selecoes = [resultado.columns.tolist()] + resultado.values.tolist()

#Agora vou correr o dataframe linha por linha e se estiver dentro da condição ele vai adicionar a linha na minha lista
for _, row in df.iterrows():
    custo_investimento = row['Custo do investimento (R$)']
    # Checando se o custo total mais o custo do investivento da linha que está sendo lida é menor que 2400000
    if custo_total + custo_investimento <= capital_para_investimento:
      #agora o programa vai ver se o risco medio, alto e baixo já não passaram do valor máximo para cada categoria
        soma_risco_medio = sum(item[2] for item in selecoes if item[4] == 'Médio')
        soma_risco_alto = sum(item[2] for item in selecoes if item[4] == 'Alto')
        soma_risco_baixo = sum(item[2] for item in selecoes if item[4] == 'Baixo')
        if soma_risco_medio >= valor_max_investimento_medio:
            continue
        elif soma_risco_alto >= valor_max_investimento_alto:
            continue
        elif soma_risco_baixo >= valor_max_investimento_baixo:
            continue
        else:
            selecoes.append(row)
            custo_total += custo_investimento
    else:
        continue

# Criando um novo dataframe com a lista preenchida pelos investimentos.
df_selecoes = pd.DataFrame(selecoes[1:], columns=selecoes[0])
# colocando em ordem de melhor investimento e mostrando o valor total que será necessário investir.
df_selecoes = df_selecoes.sort_values(by='Melhores Investimentos',ascending=False)
display(df_selecoes)
Capital = df_selecoes['Custo do investimento (R$)'].sum()
print(f'Capital total investido: R$ {Capital}')
```
