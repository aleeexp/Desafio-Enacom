```python
#importando bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("bmh")
```

```python
#importando os dados
df = pd.read_csv("Tabela.csv", sep=';', encoding='utf-8')
df.head(13)
```

```python
df.loc[len(df)] = ({"Opção":"14","Descrição":"Investimento novo","Custo do investimento (R$)":350,"Retorno esperado (R$)":350,"Risco do investimento (R$)":"Alto"})
df.loc[len(df)] = ({"Opção":"15","Descrição":"Investimento com Erro","Custo do investimento (R$)":620,"Retorno esperado (R$)":500,"Risco do investimento (R$)":"Alto"})
df.loc[len(df)] = ({"Opção":"16","Descrição":"Investimento com risco diferente","Custo do investimento (R$)":250,"Retorno esperado (R$)":350,"Risco do investimento (R$)":"Alto"})
```

```python
df['Porcentagem de ganho (%)'] = df['Retorno esperado (R$)'] / df['Custo do investimento (R$)'] *100
df = df.sort_values(by='Retorno esperado (R$)',ascending=False)
df
```

```python
df['Retorno X Porcentagem'] = df['Retorno esperado (R$)'] * df['Porcentagem de ganho (%)']
df = df.sort_values(by='Retorno X Porcentagem',ascending=False)
df
```

```python
# Inicializar as variáveis para armazenar as seleções
selecoes = []
custo_total = 0
riscos_selecionados = []

# Percorrer o DataFrame
for _, row in df.iterrows():
    risco = row['Risco do investimento (R$)']
    custo_investimento = row['Custo do investimento (R$)']

    # Verificar se a opção atende aos critérios de risco e custo
    if riscos_selecionados.count(risco) < 3 and custo_total + custo_investimento <= 2400:
        riscos_selecionados.append(risco)
        selecoes.append(row)
        custo_total += custo_investimento


    # Verificar se todas as opções de risco foram selecionadas duas vezes e se o custo total ultrapassa 2400
    if riscos_selecionados.count('Médio') >= 2 and riscos_selecionados.count('Alto') >= 1 and riscos_selecionados.count('Baixo') >= 2 and custo_total >= 2400:
        break

# Criar um novo DataFrame com as seleções
df_selecoes = pd.DataFrame(selecoes)

# Imprimir as seleções
display(df_selecoes)


```

```python
# Inicializar as variáveis para armazenar as seleções
selecoes = []
custo_total = 0
risco_medio = 0
risco_alto = 0
risco_baixo = 0

# Percorrer o DataFrame
for _, row in df.iterrows():
    risco = row['Risco do investimento (R$)']
    custo_investimento = row['Custo do investimento (R$)']

    # Verificar se a opção atende aos critérios de risco e custo
    if (
        (risco == 'Médio' and risco_medio < 2) or
        (risco == 'Alto' and risco_alto < 1) or
        (risco == 'Baixo' and risco_baixo < 2)
    ) and custo_total + custo_investimento <= 2400:
        selecoes.append(row)
        custo_total += custo_investimento
        if risco == 'Médio':
            risco_medio += 1
        elif risco == 'Alto':
            risco_alto += 1
        elif risco == 'Baixo':
            risco_baixo += 1
        else:
            custo_total -= custo_investimento
            selecoes.pop()

    elif (
        (risco == 'Médio' and risco_medio < 2) or
        (risco == 'Alto' and risco_alto < 1) or
        (risco == 'Baixo' and risco_baixo < 2)
    ) and custo_total + custo_investimento > 2400:
        custo_total += custo_investimento


    elif ( 
        (risco == 'Médio' and risco_medio > 1) or
        (risco == 'Alto' and risco_alto > 0) or
        (risco == 'Baixo' and risco_baixo > 1)
    ) and custo_total + custo_investimento <= 2400:
        selecoes.append(row)
        custo_total += custo_investimento
        if risco == 'Médio':
            risco_medio += 1
        elif risco == 'Alto':
            risco_alto += 1
        elif risco == 'Baixo':
            risco_baixo += 1
        else:
            custo_total -= custo_investimento
            selecoes.pop()

    else:
        selecoes.append(row)
        selecoes.pop()



# Criar um novo DataFrame com as seleções
df_selecoes = pd.DataFrame(selecoes)

# Imprimir as seleções
display(df_selecoes)
```
